# **Write a Pandas program to implement following operation**

```python
# Import necessary libraries
import numpy as np
import pandas as pd 
```

* **to convert a NumPy array to a Pandas series**

```python
arr = np.arange(1,10)
print("Numpy array is :", arr) 

series = pd.Series(arr)
print("\nPandas Series : \n",series)
```

* **to convert the first column of a DataFrame as a Series**

```python
dic = {
    "name" : ['abc', 'pqr', 'xyz'],
    "marks" : [99, 93, 90],
    "city" : ['rajkot', 'porbandar', 'junagadh']
}
df = pd.DataFrame(dic)
df
```

```python
s = pd.Series(df['name'])
print("\n1st column as a Series : \n",s)
```

* **to create the mean and standard deviation of the data of a given Series**

```python
print("Original Data Series : \n", series)
print("\nMean of the Data Series :", series.mean())
print("\nStandard deviation of the Data Series :", series.std())
```

* **to sort a given Series**

```python
print("Series before sorting : \n", series)
print("\nSeries after sorting : \n",series.sort_values())
```
