# **Write a Pandas program to create a line plot of the opening, closing stock prices of given company between two specific dates**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('/kaggle/input/national-stock-exchange-time-series/nifty_it_index.csv', parse_dates=['Date'])
```

```python
df.set_index('Date', inplace=True)

plt.figure(figsize=(12, 6))

plt.plot(df.index, df['Open'], label='Open', color='green')
plt.plot(df.index, df['Close'], label='Close', color='red')

plt.legend(loc='best')

plt.title('Nifty IT Index: Open vs. Close Prices')
plt.xlabel('Date')
plt.ylabel('Price')

plt.show()
```
