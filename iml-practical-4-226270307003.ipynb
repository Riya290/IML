# Write a NumPy program to implement following operation

```python
# Import necessary libraries
import numpy as np
```

* **add, subtract, multiply, divide arguments element-wise**

```python
# Create an array of eight values evenly spaced between 0 and 100
x = np.linspace(0, 100, 8, dtype="float32")
print("x : \n", x)

# Addition +
print("\nnp.add(x, 2):\n", np.add(x, 2))  

# Subtraction -
print("\nnp.subtract(x, 2):\n", np.subtract(x, 2)) 

# Multiplication *
print("\nnp.multiply(x, 2):\n", np.multiply(x, 2))

# Division /
print("\nnp.divide(x, 2):\n", np.divide(x, 2))              
```

* **round elements of the array to the nearest integer**

```python
print("\nnp.round(x):\n", np.round(x))
```

* **calculate mean across dimension, in a 2D numpy array**

```python
# Creating a 2x2 NumPy array
y = np.array([[18, 29], [5, 23]])
print("\nOriginal array : \n", y)

# Computing the mean of each column using axis 0 (column-wise)
print("\nMean of each column : \n", y.mean(axis=0))

# Computing the mean of each row using axis 1 (row-wise)
print("\nMean of each row : \n", y.mean(axis=1))
```

* **calculate the difference between neighboring elements, element-wise of a given array**

```python
print("\nDifference between neighboring elements, element-wise of the y array : \n", np.diff(y))
```
