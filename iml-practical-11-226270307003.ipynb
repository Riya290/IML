# **Write a Pandas program to filter all columns where all entries present, check which rows and columns has a NaN and finally drop rows with any NaNs from the given dataset.**

```python
# Import necessary libraries
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/kaggle/input/national-stock-exchange-time-series/nifty_it_index.csv')

print("Original DataFrame:")
df
```

```python
df.loc[[0, 10, 150, 188, 245, 23], ['Close']] = np.nan
df.loc[[18, 12, 29, 150, 131, 117], ['Volume']] = np.nan
```

```python
nan_value = df.isna()
print("\nLocations of NaN values: \n", nan_value)
```

```python
df = df.dropna()
print("\nDataFrame after dropping rows with any NaN values: \n", df)
```
